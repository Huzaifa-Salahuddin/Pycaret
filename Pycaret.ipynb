```python
import pandas as pd
df = pd.read_csv('day.csv')
df.head(1)
```

```python
from pycaret.regression import setup, compare_models, evaluate_model, predict_model, save_model, load_model
```

```python
Test = setup(df, target='cnt')
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
test_data = pd.read_csv('day_test.csv')
predict_model(best_model, data=test_data)
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
save_model(best_model, 'linear')
```

# PyCaret For Classification

```python
df = pd.read_csv('train.csv')
df.head(1)
```

```python
from pycaret.classification import setup, compare_models, evaluate_model, predict_model, save_model, load_model
```

```python
classification = setup(df, target='Survived')
```

```python
best_model_classification = compare_models()
```

```python
evaluate_model(best_model_classification)
```

```python
test = pd.read_csv('test.csv')
predict_model(best_model_classification, data=test)
```

```python
best_model_classification
```

```python
save_model(best_model_classification, 'logistic')
```
